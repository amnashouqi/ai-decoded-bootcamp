**Load the dataset**

```python
import pandas as pd

url = 'https://raw.githubusercontent.com/amnashouqi/ai-decoded-bootcamp/main/housing.csv'
df = pd.read_csv(url)

df.head()
```


**Look at the columns**


```python
print("Shape (rows, columns):", df.shape)
print("\nColumn names:")
print(df.columns.tolist())

print("\nData types:")
df.info()
```

**Check for missing values**

```python
df.isnull().sum()
```

**Basic stats**

```python
df.describe()
```

**Plot distributions (histograms)**

```python
import matplotlib.pyplot as plt

df[['median_income', 'housing_median_age', 'median_house_value']].hist(
    bins=30, figsize=(12, 6)
)
plt.tight_layout()
plt.show()
```

**Categorical column distribution**

```python
df['ocean_proximity'].value_counts().plot(kind='bar', figsize=(6,4))
plt.title('Number of blocks by ocean proximity')
plt.ylabel('Count')
plt.show()
```

**Scatter plot: income vs. house value**

```python
df.plot(kind='scatter', x='median_income', y='median_house_value',
        alpha=0.3, figsize=(8,6))
plt.title('Median Income vs Median House Value')
plt.show()
```
